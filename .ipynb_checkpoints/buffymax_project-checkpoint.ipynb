# Explore Our Datasets

```python
# import the relevant packages
import ast
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set matplotlib inline
%matplotlib inline
from datetime import datetime
import nltk
# nltk.download("stopwords")
# nltk.download("words")
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords, words
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
s = PorterStemmer()
```

```python
def convert_timestamp(x):
    return datetime.fromtimestamp(x)

lemmatizer = WordNetLemmatizer()

def lemmatize(pair):
    word, tag = pair
    try:
        return lemmatizer.lemmatize(word, pos=tag[0].lower())
    except KeyError:
        return word
    
def cast_getter(cast_list):
    cast_list = ast.literal_eval(cast_list)
    gernerated_list = [d.get('name') for d in cast_list]
    return gernerated_list
```

### Box Office Dataset

```python
box_office_df = pd.read_csv('boxoffice_march_2019.csv')
box_office_df.head()
```

```python
print(len(box_office_df['title'].unique()))
print(len(box_office_df))
```

```python
unique_title = []
for title in box_office_df['title'].unique():
    unique_title.append(title)
print(unique_title[:10])
```

```python
print(box_office_df['title'].value_counts().nlargest(10))
```

```python
box_office_df[box_office_df['title'] == 'Bluebeard']
```

```python
box_office_df[box_office_df['title'] == 'Heaven']
```

```python
box_office_df['lifetime_in_millions'] = box_office_df['lifetime_gross'] / 1000000
box_office_df['lifetime_in_millions'].describe()
```

```python
f, axes = plt.subplots(1, 1, figsize = (17,3))
sns.boxplot(data = box_office_df['lifetime_in_millions'], orient = "h", color = 'green')
```

```python
p25_revenue = np.percentile(box_office_df['lifetime_in_millions'], 25)
p75_revenue = np.percentile(box_office_df['lifetime_in_millions'], 75)
IQR = p75_revenue - p25_revenue
step_size = 1.5 * IQR
lower_step = p25_revenue - step_size
upper_step = p75_revenue + step_size

outlier_wine_upper = box_office_df[box_office_df['lifetime_in_millions'] > upper_step]
print("The no. of outliers above is", len(outlier_wine_upper))
outlier_wine_lower = box_office_df[box_office_df['lifetime_in_millions'] < lower_step]
print("The no. of outliers is", len(outlier_wine_lower))
```

```python
top_MillionRevenue = box_office_df.nlargest(20, 'lifetime_in_millions')
```

```python
dims = (16, 5)
fig = plt.figure(figsize = dims)

ax1 = fig.add_subplot(111)

top_MillionRevenue.plot(kind = 'bar', x = 'title', y = 'lifetime_in_millions', color = 'red', ax = ax1)
plt.title("Top 10 Box Office Figures")
plt.ylabel("Box Office Revenue (in millions)")
ax1.set_xlabel('Movie Titles')
```

```python
top_MillionRevenue_df = box_office_df.sort_values(['lifetime_in_millions'], ascending = False)
top_MillionRevenue_df.head(10)
```

```python
credits_df = pd.read_csv('credits.csv')
credits_df.head()
```

```python
metadata_df = pd.read_csv('movies_metadata.csv')
metadata_df.head(2)
```

```python
print("The original dataset columns:\n", list(metadata_df.columns.values))
improve_metadata_df = metadata_df.drop(columns = ['belongs_to_collection', 'genres', 'homepage', 'imdb_id',
                                                  'overview', 'spoken_languages', 'tagline', 'production_companies',
                                                  'production_countries', 'poster_path', 'status'], axis = 1)
improve_metadata_df = improve_metadata_df.dropna()
improve_metadata_df.head()
```
